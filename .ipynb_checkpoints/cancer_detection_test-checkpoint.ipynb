```python
import cv2
import matplotlib.pyplot as plt
import _pickle as cPickle
import numpy as np

f = open("/home/jtlee/workspace/cancer_segmentation/ckpt/ckpt_10.pkl", "rb")
rf = cPickle.load(f)

```

```python
# Read image
target_name = "11.jpg"
target_file = "/home/jtlee/workspace/color-recognition-v1/workspace/temp/" + target_name
image = cv2.imread(target_file)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure()
plt.axis("off")
plt.imshow(image)
plt.show()
```

```python
# Skin detection
img_hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
flat = img_hls.reshape((-1, 3))

#%timeit pred = rf.predict(flat)
pred = rf.predict(flat)
```

```python
# masking
mask = np.full([pred.shape[0], 3], 0, dtype = 'uint8')
indices = np.where(pred == 1)
mask[indices] = image.reshape((-1, 3))[indices]
mask = mask.reshape(list(image.shape))
mask2 = np.full([pred.shape[0], 3], 0, dtype = 'uint8')
indices2 = np.where(pred == 0)
mask2[indices2] = image.reshape((-1, 3))[indices2]
mask2 = mask2.reshape(list(image.shape))

plt.figure()
plt.axis("off")
plt.imshow(mask)
plt.savefig("skin_"+target_name)
plt.show()

plt.figure()
plt.axis("off")
plt.imshow(mask2)
plt.savefig("skin_"+target_name)
plt.show()

plt.figure()
plt.axis("off")
plt.imshow(mask+mask2)
plt.savefig("skin_"+target_name)
plt.show()
```
